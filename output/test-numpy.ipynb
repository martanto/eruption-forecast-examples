```python
import numpy as np
import numpy.ma as ma
```

```python
test = np.array([0,0,1,1,1,1])
```

```python
np.mean(test)
```

```python
test_masked = ma.masked_equal(test,0)
```

```python
test_masked
```

```python
test_masked = ma.masked_values(test_masked, 1.0)
```

```python
test_masked
```

```python
np.sum(test_masked) is ma.masked
```
