```python
LOW_BAND  = (0.5, 5)
HIGH_BAND = (5, 15)
```

```python
def frequency_index(x, fs):
    x = np.reshape(x, (1024, ))
    spec = np.abs(rfft(x))**2
    freqs = rfftfreq(len(x), 1/fs)
    El = np.sum(spec[(freqs >= LOW_BAND[0]) & (freqs <= LOW_BAND[1])])
    Eh = np.sum(spec[(freqs >= HIGH_BAND[0]) & (freqs <= HIGH_BAND[1])])
    return np.log10(Eh / El) if El > 0 else np.nan
```

```python
fi = frequency_index(data, 100)
```
