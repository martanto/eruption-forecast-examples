```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
from glob import glob
from datetime import datetime
import os
```

```python
eruptions = [
    "2025-03-20",
    "2025-04-22",
    "2025-05-18",
    "2025-06-17",
    "2025-07-07",
    "2025-08-01",
    "2025-08-17",
]
```

```python
files = glob(
    os.path.join(
        r"D:\Projects\eruption-forecast\output\forecast\VG.OJN.00.EHN\tremor\_tmp\*.csv"
    )
)
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file, index_col=0, parse_dates=True)
    dfs.append(df)
```

```python
df = pd.concat(dfs)
```

```python
columns = df.columns
fig, axs = plt.subplots(nrows=len(columns), ncols=1, figsize=(10, 2.5 * len(columns)))
for index, column in enumerate(columns):
    ax = axs[index]
    ax.plot(df.index, df[column], color="black", linewidth=1, label=f"{column}")
    ax.tick_params(labelsize=8)
    ax.set_xlim(df.index[0], df.index[-1])

    for eruption in eruptions:
        ax.axvline(datetime.strptime(eruption, "%Y-%m-%d"), color="red", linestyle="--")

    ax.xaxis.set_major_locator(mdates.DayLocator(interval=14))
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
    ax.legend(loc="upper left", ncol=4, fontsize=6, frameon=False)
    for label in ax.get_xticklabels(which="major"):
        label.set(rotation=15, horizontalalignment="right")
plt.tight_layout()
```
