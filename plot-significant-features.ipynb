```python
import os
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
sns.color_palette("flare")
sns.set_style("white")
sns.set_context("paper")
```

```python
features = r"D:\Projects\eruption-forecast\output\VG.OJN.00.EHZ\trainings\all_features\00000.csv"
```

```python
df = pd.read_csv(features)
# df = df.iloc[::-1]
top_features = 20
number_of_features = 50
title = None
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(3, 12))
ax.grid(True, linestyle="-.", axis="both", alpha=0.3)
sns.barplot(x="values", y="features", data=df.iloc[0:49], width=0.5)
# ax.barh(df["features"], df["values"], height=0.5)
ax.axhline(
    df.index[9],
    color="orange",
    linestyle="-.",
    label=f"Top 10 Fts",
)
ax.axhline(
    df.index[top_features - 1],
    color="red",
    linestyle="-.",
    label=f"Top {top_features} Fts",
)
ax.title.set_text(title or f"{number_of_features} Significant Features")
ax.legend(frameon=False)
ax.set_ylabel(None)
ax.set_xlabel(None)
# plt.ylim(-0.5, number_of_features - 0.5)
```
