```python
def to_series(df: pd.DataFrame, column_value: str, column_index: str = "id") -> pd.Series:
    series = pd.Series(df[column_value])
    series.index = labels[column_index]
    return series
```

```python
series = to_series(labels, column_value="is_erupted")
```
