```python
from eruption_forecast.decorators import save_properties
```

```python
@save_properties('outputs/user_config.json', save_as='json')
class UserConfig:
    """User configuration with auto-save capability"""

    def __init__(self, username: str, email: str):
        self.username = username
        self.email = email
        self.preferences = {
            'theme': 'dark',
            'notifications': True
        }
        self.login_count = 0

    def update_preferences(self, **kwargs):
        self.preferences.update(kwargs)
        self.save_properties()

    def increment_login(self):
        self.login_count += 1
```

```python
user = UserConfig("john_doe", "john@example.com")
user.update_preferences(theme='light', language='en')
user.increment_login()
saved_path = user.save_properties()
print(f"✓ Properties saved to: {saved_path}\n")
```
