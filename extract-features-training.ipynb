```python
def get_target(training_label_csv: str) -> pd.Series:
    df_label = pd.read_csv(training_label_csv)
	df_label["is_erupted"] = df_label["is_erupted"].astype(bool)
	target = pd.Series(df_label["is_erupted"].values, index=df_label["id"].values)
	return target
```
