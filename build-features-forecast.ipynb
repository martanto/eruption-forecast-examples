```python
import pandas as pd
import os
from glob import glob
from datetime import datetime, timedelta
from eruption_forecast import LabelBuilder
from eruption_forecast.utils import construct_windows
```

```python
forecast_start_date = "2025-08-10"
forecast_end_date = "2025-08-18"
window_step = 10
window_step_unit = "minutes"
forecast_start_date_obj = datetime.strptime(forecast_start_date, "%Y-%m-%d")
forecast_end_date_obj = datetime.strptime(forecast_end_date, "%Y-%m-%d")
```

```python
forecast_dir = r"D:\Projects\eruption-forecast\examples\output\VG.OJN.00.EHN\forecast"
features_matrix_dir = os.path.join(forecast_dir, "features_matrix")
os.makedirs(forecast_dir, exist_ok=True)
```

```python
label_forecast = f"forecast_window_{forecast_start_date}-{forecast_end_date}_step-{window_step}-{window_step_unit}"
label_forecast_csv = os.path.join(forecast_dir, f"{label_forecast}.csv")
```

```python
def build_forecast_window(forecast_csv: str) -> pd.DataFrame:
	if os.path.exists(forecast_csv):
		return pd.read_csv(forecast_csv, index_col=0, parse_dates=True)

	df_forecast_windows = construct_windows(
		window_step=window_step,
		window_step_unit=window_step_unit,
		start_date=forecast_start_date,
		end_date=forecast_end_date,
	)

	df_forecast_windows["id"] = range(len(df_forecast_windows))
	df_forecast_windows.to_csv(label_forecast_csv, index=True)
	return df_forecast_windows
```
